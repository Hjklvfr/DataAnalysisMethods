## Restore borough nas

Is it possible to restore 449210 na borough values by lat and lon using GIS data?

Use https://data.cityofnewyork.us/City-Government/Borough-Boundaries/tqmj-j8zm#revert data for this task

```python
df['borough'].unique()  # check existing name format of boroughs
```

```python
from analytics.src.lab2 import borough_by_coords
import mapply
from shapely.geometry import Point

mapply.init(n_workers=4)

boroughs = borough_na_with_coords['location'].apply(lambda x: Point(float(eval(x)[1]), float(eval(x)[0])))
boroughs = boroughs.mapply(lambda x, b_func:
                                  b_func(x), args=(borough_by_coords,))
```

```python
df
```

```python
df = df.copy()
df['borough'] = df['borough'].fillna(boroughs).copy()
```

```python
df.isna().sum()
```

## Look at accident count

```python
month_count_df = df.groupby([pd.Grouper(key='crash_datetime', freq='ME'), 'borough']).agg(len)[['collision_id']].rename({'collision_id': 'count'}, axis=1)
month_count_df
```

```python
sns.lineplot(month_count_df, x='crash_datetime', y='count', hue='borough', dashes=False, palette=sns.color_palette('tab10', n_colors=6))
```
