recursion. They must retrieve page contents for a URL, examine that page for another URL, and retrieve
that page, ad infinitum.
The recommended Python library for this assignment is Scrapy. Using other libraries like pandas and
matplotlib is allowed.
More about Scapy crawlers: https://doc.scrapy.org/en/latest/topics/spiders.html#crawlspider
Working process:
1. Install Scrapy.
2. Extract data from Steam (https://store.steampowered.com/). Using CrawlSpider from Scrapy get the
following info: Game title, release date, developer, publisher, popular user-defined tags for a product
from first 1000 products of “Top Sellers”.

<br>
**Advanced task:** to get all the whole tags list available by “+” button.
Note: For extracting tags, you can use CSS and XPath selectors.
3. Group your games dataset by year. Name 3 the most popular game tags in each year. Name the least
popular game tags in each year.
4. For these tags, print game names that have such tags (considering the proper year).
Note: e.g., if tag “Open World” was popular in 2020, but was not in a top in 2019, only game titles with
tag “Open World” from 2020 should be printed.

```python
import os
from modules.vkapi import VkApi
import pandas as pd
import seaborn as sns
import pymongo

from dotenv import load_dotenv

load_dotenv()
```

```python
client = pymongo.MongoClient(os.getenv("MONGO_URL"))
products = client['steam']['products']

df = pd.DataFrame(list(products.find({})))
df
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df
```

```python
df['year'] = df['release_date'].dt.year
df
```

```python
df.explode('tags')
```

```python
tag_counts = df.explode('tags').groupby(['year', 'tags']).size().reset_index(name='count')
tag_counts
```

```python
tag_counts.loc[tag_counts.groupby('year')['count'].idxmax()]
```

```python
top_tag_by_year = tag_counts.loc[tag_counts.groupby('year')['count'].idxmax()]
```

```python
top_tag_by_year['games'] = top_tag_by_year.apply(
    lambda row: df[
        df['tags'].apply(lambda x: row['tags'] in x if isinstance(x, list) else False)
    ]['title'].head(10).tolist(),
    axis=1
)
```
